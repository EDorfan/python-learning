```python
import json

people_string = '''
{
    "people": [
        {
            "name": "John Smith",
            "phone": "615-555-7164",
            "emails": ["john.doe@example.com", "john@doe.com"],
            "has_license": false
        },
        {
            "name": "Jane Doe",
            "phone": "615-555-7165",
            "emails": ["jane.doe@example.com", "jane@doe.com"],
            "has_license": true
        }
    ]
}
'''

# Parse the JSON string into a Python object
data = json.loads(people_string)

# Data type is a dictionary
print(type(data))

print(data)

# Access the list of people
print(data["people"])

for person in data['people']:
    print(person)
    print(person['name'])
    print(person['phone'])
    print(person['emails'])
    print(person['has_license'])

# Convert Python object back to JSON string
# Delete 1 element to show how to use the index

for person in data['people']:
    del person['phone']

# dumps() is used to convert a Python object to a JSON string
# indent is used to indent the JSON string
# sort_keys is used to sort the keys of the JSON string
new_string = json.dumps(data, indent=2, sort_keys=True)

print(new_string)

# Convert Python object back to JSON string
# Delete 1 element to show how to use the index

```
