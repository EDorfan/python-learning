## What is a list?

- A list is an ordered, mutable sequence.
- Lists preserve insertion order.
- Lists are indexable/sliceable and can hold mixed types.
- Create with `[]` or `list(iterable)`.


```python
# Basics: creation, indexing, slicing, mutability
empty = []
numbers = [10, 20, 30, 40]
mixed = [1, "two", 3.0, [4, 5]]
from_iterable = list(range(3))  # [0, 1, 2]

print("Creation:", empty, numbers, mixed, from_iterable)

# Indexing
print("Indexing:", numbers[0], numbers[-1])

# Slicing
print("Slicing:", numbers[1:3], numbers[:], numbers[::-1])

# Mutability
nums = numbers[:]
nums[1] = 25
nums.append(50)
nums.extend([60, 70])
nums.insert(0, 5)
removed = nums.pop()

# If you remove an item which is not in the list, you get the following error:
# x not in list
nums.remove(25)

# Can concatenate lists with +
nums = nums + [80, 90]
mixed_2 = numbers + mixed

print("Mutated list:", nums, "removed=", removed)
```

## Useful methods


```python
# count, index, sort, reverse, clear
data = [3, 1, 2, 3, 2, 3]
print("Count of 3:", data.count(3))
print("Index of first 2:", data.index(2))

copy_for_sort = data[:]
copy_for_sort.sort()

# Can also use the sorted function to return a new list
sorted_data = sorted(data)

print("Sorted in place:", copy_for_sort)
print("Sorted (new list):", sorted_data)

descending = sorted(data, reverse=True)
print("Sorted (new list, desc):", descending)

words = ["pear", "apple", "banana", "fig"]
words_sorted_by_length = sorted(words, key=len)
print("Sorted by length:", words_sorted_by_length)

# Reverses the list
words.reverse()
print("Reversed:", words)

tmp = [1, 2, 3]
tmp.clear()
print("Cleared list:", tmp)
```

## Copying vs aliasing


```python
from copy import deepcopy

original = [[1, 2], [3, 4]]
alias = original  # no copy
shallow = original[:]  # shallow copy (top-level only)
deep = deepcopy(original)  # deep copy

alias[0][0] = 999
print("After alias mutation:")
print("original:", original)
print("shallow:", shallow)
print("deep:", deep)
```

## List comprehensions


```python
squares = [n * n for n in range(6)]
evens = [n for n in range(10) if n % 2 == 0]
cartesian_pairs = [(x, y) for x in range(2) for y in range(3)]
print("Comprehensions:", squares, evens, cartesian_pairs)
```

## Edge cases and pitfalls


```python
from typing import List

# Empty list operations
empty: List[int] = []
print("Empty length:", len(empty))
print("Membership on empty:", 1 in empty)
# Accessing empty[0] would raise IndexError
# empty[0]

# Negative indices
nums = [1, 2, 3]
print("Negative index -1:", nums[-1])

# Slices are forgiving
print("Slice beyond bounds:", nums[10:20])

# Removing a non-existent item raises ValueError
# nums.remove(999)

# Sorting with incomparable mixed types raises TypeError
mixed = [1, 2, "three"]
# sorted(mixed)
print("Mixed types present (sorting would fail):", mixed)

# Pitfall: list multiplication with nested lists (aliasing rows)
bad_grid = [[0] * 3] * 3
bad_grid[0][0] = 1
print("Bad grid (aliasing):", bad_grid)

good_grid = [[0 for _ in range(3)] for _ in range(3)]
good_grid[0][0] = 1
print("Good grid (independent rows):", good_grid)
```

## Broken example to fix

`remove_all_bad` tries to remove all occurrences but mutates while iterating forward, causing skips. Compare with a correct approach.


```python
def remove_all_bad(items: list[int], value: int) -> list[int]:
    for i, x in enumerate(items):
        if x == value:
            items.pop(i)  # BUG: index shifts left, next element skipped
    return items

# Demonstrate the bug
print(remove_all_bad([1, 2, 2, 2, 3], 2))  # Likely [1, 2, 3]
```

```python
# Correct approach

def remove_all_good(items: list[int], value: int) -> list[int]:
    return [x for x in items if x != value]

print(remove_all_good([1, 2, 2, 2, 3], 2))  # [1, 3]
```
